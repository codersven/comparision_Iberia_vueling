```python
import pymongo

def get_medical_top_spots(provincia="Badajoz"):
    client = pymongo.MongoClient("mongodb://localhost:27017/")
    db = client["Espana"]
    
    # Suche Orte mit Krankenhaus-Eintrag
    query = {
        "Provincia": provincia,
        "Piscina": {"$ne": ""},
        "Biblioteca": {"$ne": ""},
        "Hospital": {"$ne": ""}
    }
    
    projection = {
        "Ciudad": 1, 
        "Hospital": 1, 
        "Farmacia": 1, 
        "Poplación": 1,
        "Piscina":1,
        "Biblioteca":1,
        "_id": 0
    }
    
    # Wir laden die Daten
    top_spots = list(db.ciudad.find(query, projection).sort("Poplación", -1))
    
    print(f"--- Medizinische Hotspots in {provincia} ---")
    for spot in top_spots:
        # Sicherer Zugriff mit .get(key, default_value)
        stadt = spot.get("Ciudad", "Unbekannt")
        pop = spot.get("Poplación", 0)
        hospital = spot.get("Hospital", "Keine Info")
        farmacia = spot.get("Farmacia", "Keine Info")
        piscina = spot.get("Piscina", "Keine Info")
        biblioteca = spot.get("Biblioteca", "Keine Info")
        
        print(f"Stadt: {stadt} ({pop} Einw.)")
        print(f"🏥: {hospital[:60]}...") 
        print(f"💊: {farmacia}")
        print(f"👙: {piscina}")
        print(f"📚: {biblioteca}")
        print("-" * 30)

get_medical_top_spots()
```
