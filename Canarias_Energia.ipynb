```python
from pyjstat import pyjstat
import pandas as pd

file_path = 'Energía_eléctrica_según_producción_bruta_demanda_interior_y_consumo_final_Islas_de_Canarias_por_periodos.jsonstat'

# Datei einlesen
with open(file_path, 'r', encoding='utf-8') as f:
    js_data = f.read()

# Datensatz parsen
dataset = pyjstat.Dataset.read(js_data)

# Umwandeln in DataFrame mit der neuen Syntax
df = dataset.write(output='dataframe')

# Erfolgskontrolle
print(df.head())
```
