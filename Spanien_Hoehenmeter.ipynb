```python
import pandas as pd
from pymongo import MongoClient

# Verbindung zur Datenbank
client = MongoClient("mongodb://localhost:27017/")
db = client["Espana"]
usa_col = db["ciudad"]

# Daten laden
cursor = usa_col.find({}, {"Ciudad": 1, "Provincia": 1, "Población": 1, "Densidad": 1, "Altitud": 1})
df = pd.DataFrame(list(cursor))

# --- NEU: Pandas anweisen, ALLES anzuzeigen ---
pd.set_option('display.max_rows', None)  # Keine Begrenzung bei den Zeilen
pd.set_option('display.max_columns', None) # Keine Begrenzung bei den Spalten
pd.set_option('display.width', 1000)      # Zeilenumbruch im Terminal verhindern

# Analyse: Nach Höhe sortieren
df_sorted = df.sort_values(by="Altitud", ascending=False)

print(f"Höhenlagen meiner Espana-Collection ({len(df)} Orte):")
print(df_sorted[["Ciudad", "Provincia", "Altitud"]])

# Optionale Reinigung: Falls du später wieder zum Standard zurück willst
# pd.reset_option('display.max_rows')
```

```python
import pandas as pd
from pymongo import MongoClient

# Verbindung zur Datenbank
client = MongoClient("mongodb://localhost:27017/")
db = client["Espana"]
collection = db["ciudad"]

# Die Pipeline: Filtern auf Andalusien und Sortieren nach Höhe
pipeline = [
    {
        "$match": {
            "Com autónoma": "Andalucía",
            "Altitud": { "$exists": True }
        }
    },
    {
        "$project": {
            "_id": 0,
            "Ciudad": 1,
            "Provincia": 1,
            "Altitud": 1,
            "Población": 1
        }
    },
    { "$sort": { "Altitud": -1 } }
]

# Ergebnisse abrufen
results = list(collection.aggregate(pipeline))

# In Pandas DataFrame umwandeln
df = pd.DataFrame(results)

# Pandas-Anzeigeoptionen für die Pause
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(f"Höhenlagen in Andalusien ({len(df)} Orte gefunden):")
print("-" * 70)
if not df.empty:
    print(df[["Ciudad", "Provincia", "Altitud"]])
else:
    print("Keine Daten gefunden. Prüfe den Feldnamen 'Com autónoma'.")
```
